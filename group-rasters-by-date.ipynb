```python
import glob
```

```python
# load files
rasterlist = glob.glob('./data/*.nc')
len(rasterlist) # there are 244 files
```

```python
# find unique dates and put them in a list
dates = []
for raster in rasterlist:
    raster_filename = raster.split('/')[-1]
    date = raster_filename[0:8]
    dates.append(date)

print(len(dates))                 # there are 244 files
unique_dates = list(set(dates))
print(len(unique_dates))          # there are 122 unique dates
```

```python
# group files by date (e.g.: )
# grouped_by_date = {
#     20190101: ['20190101_day_file.nc','20190101_night_file.nc'],
#     20190102: ['20190102_day_file.nc','20190102_night_file.nc'],
#     ... and so on
# }
# 
# 
grouped_by_date = {date: [] for date in unique_dates}

for date in grouped_by_date.keys():
    grouped_by_date[date]= list(filter(lambda file: date in file, rasterlist))
```

```python
# for each date, load 2 files and do your calculations in this for loop
for date, files in grouped_by_date.items():
    print('----')
    print('date is: ', date)
    print('file1 is: ', files[0])
    print('file2 is: ', files[1])
```
